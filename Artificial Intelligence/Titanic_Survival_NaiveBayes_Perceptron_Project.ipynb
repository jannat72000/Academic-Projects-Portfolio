```python
proba = {}

def count_probab():
    # Calculate probabilities for the 'Sex' column
    male_0 = NW_data[(NW_data["Sex"] == "male") & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["male_0"] = male_0

    male_1 = NW_data[(NW_data["Sex"] == "male") & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["male_1"] = male_1

    female_0 = NW_data[(NW_data["Sex"] == "female") & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["female_0"] = female_0

    female_1 = NW_data[(NW_data["Sex"] == "female") & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["female_1"] = female_1

    # Calculate probabilities for the 'Pclass' column
    pclass_1_0 = NW_data[(NW_data["Pclass"] == 1) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["pclass_1_0"] = pclass_1_0

    pclass_1_1 = NW_data[(NW_data["Pclass"] == 1) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["pclass_1_1"] = pclass_1_1

    pclass_2_0 = NW_data[(NW_data["Pclass"] == 2) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["pclass_2_0"] = pclass_2_0

    pclass_2_1 = NW_data[(NW_data["Pclass"] == 2) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["pclass_2_1"] = pclass_2_1

    pclass_3_0 = NW_data[(NW_data["Pclass"] == 3) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["pclass_3_0"] = pclass_3_0

    pclass_3_1 = NW_data[(NW_data["Pclass"] == 3) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["pclass_3_1"] = pclass_3_1

    # Calculate probabilities for the 'Age' column (assuming normal distribution)
    age_mean_not = NW_data[NW_data['Survived'] == 0]['Age'].mean()
    age_std_not = NW_data[NW_data['Survived'] == 0]['Age'].std()
    age_prob_not = 1 / (age_std_not * np.sqrt(2 * np.pi)) * np.exp(- (NW_data["Age"] - age_mean_not)**2 / (2 * age_std_not**2))
    proba["age_0"] = age_prob_not

    age_mean = NW_data[NW_data['Survived'] == 1]['Age'].mean()
    age_std = NW_data[NW_data['Survived'] == 1]['Age'].std()
    age_prob = 1 / (age_std * np.sqrt(2 * np.pi)) * np.exp(- (NW_data["Age"] - age_mean)**2 / (2 * age_std**2))
    proba["age_1"] = age_prob

    # Calculate probabilities for the 'SibSp' column
    sibsp_0_0 = NW_data[(NW_data["SibSp"] == 0) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["sibsp_0_0"] = sibsp_0_0

    sibsp_0_1 = NW_data[(NW_data["SibSp"] == 0) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["sibsp_0_1"] = sibsp_0_1

    sibsp_1_0 = NW_data[(NW_data["SibSp"] == 1) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["sibsp_1_0"] = sibsp_1_0

    sibsp_1_1 = NW_data[(NW_data["SibSp"] == 1) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["sibsp_1_1"] = sibsp_1_1

    # Calculate probabilities for the 'Parch' column
    parch_0_0 = NW_data[(NW_data["Parch"] == 0) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["parch_0_0"] = parch_0_0

    parch_0_1 = NW_data[(NW_data["Parch"] == 0) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["parch_0_1"] = parch_0_1

    parch_1_0 = NW_data[(NW_data["Parch"] == 1) & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["parch_1_0"] = parch_1_0

    parch_1_1 = NW_data[(NW_data["Parch"] == 1) & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["parch_1_1"] = parch_1_1


    # Calculate probabilities for the 'Embarked' column

    embarked_Q_0 = NW_data[(NW_data["Embarked"] == 'Q') & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["embarked_Q_0"] = embarked_Q_0

    embarked_Q_1 = NW_data[(NW_data["Embarked"] == 'Q') & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["embarked_Q_1"] = embarked_Q_1

    embarked_S_0 = NW_data[(NW_data["Embarked"] == 'S') & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["embarked_S_0"] = embarked_S_0

    embarked_S_1 = NW_data[(NW_data["Embarked"] == 'S') & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["embarked_S_1"] = embarked_S_1

    embarked_C_0 = NW_data[(NW_data["Embarked"] == 'C') & (NW_data["Survived"] == 0)].count()[0] / NW_data[NW_data["Survived"] == 0].count()[0]
    proba["embarked_C_0"] = embarked_C_0

    embarked_C_1 = NW_data[(NW_data["Embarked"] == 'C') & (NW_data["Survived"] == 1)].count()[0] / NW_data[NW_data["Survived"] == 1].count()[0]
    proba["embarked_C_1"] = embarked_C_1

count_probab()
```

```python
def naive_bayes_prediction(input_data):
     # Retrieve probabilities from the 'proba' dictionary based on user input
    sex_prob = proba[input_data['Sex'] + '_' + str(input_data['Survived'])]
    pclass_prob = proba['pclass_' + str(input_data['Pclass']) + '_' + str(input_data['Survived'])]
    age_prob = proba['age_' + str(input_data['Survived'])]
    sibsp_prob = proba['sibsp_' + str(input_data['SibSp']) + '_' + str(input_data['Survived'])]
    parch_prob = proba['parch_' + str(input_data['Parch']) + '_' + str(input_data['Survived'])]
    embarked_prob = proba['embarked_' + input_data['Embarked'] + '_' + str(input_data['Survived'])]

    # Calculate the overall probability using Naive Bayes assumption
    overall_prob = sex_prob * pclass_prob * age_prob * sibsp_prob * parch_prob * embarked_prob

    return overall_prob

# Example user input
input_data = {
    'Sex': 'female',
    'Pclass': 1,
    'Age': 30,
    'SibSp': 1,
    'Parch': 0,
    'Embarked': 'C',
    'Survived': 1
}

# Get the prediction using Naive Bayes
prediction = naive_bayes_prediction(input_data)
print("Prediction probability:", prediction)
```

```python
import numpy as np
import pandas as pd

# Load the Titanic dataset
data = pd.read_csv('titanic.csv')

# Extract the features and labels from the dataset
X = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']].values
y = data['Survived'].values

# Convert categorical variables into numerical values
sex_mapping = {'male': 0, 'female': 1}
embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}
X[:, 1] = [sex_mapping[val] for val in X[:, 1]]
X[:, 5] = [embarked_mapping[val] for val in X[:, 5]]

# Split the dataset into training and testing sets
split = int(0.8 * len(X))  # 80% for training, 20% for testing
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]

class Perceptron:
    def __init__(self, num_features, learning_rate=0.1, epochs=100):
        self.num_features = num_features
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = np.random.rand(num_features + 1).astype(float)  # +1 for the bias term

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        activation_input = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation(activation_input)

    def train(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                update = self.learning_rate * (y[i] - prediction)
                self.weights[1:] += update * X[i].astype(float)
                self.weights[0] += update

    def evaluate(self, X, y):
        correct_predictions = 0
        for i in range(len(X)):
            prediction = self.predict(X[i])
            if prediction == y[i]:
                correct_predictions += 1
        return correct_predictions / len(X)

# Create and train the perceptron
perceptron = Perceptron(num_features=6, learning_rate=0.1, epochs=100)

min_error = float('inf')  # Initialize minimum error with a large value
best_weights = None  # Variable to store the weights with minimum error

for epoch in range(perceptron.epochs):
    perceptron.train(X_train, y_train)
    error = 1 - perceptron.evaluate(X_train, y_train)
    print("Epoch {}: Err={:.2f}, Weights={}".format(epoch, error, perceptron.weights))

    if error < min_error:
        min_error = error
        best_weights = perceptron.weights.copy()

# Save the weight values with minimum error
np.save('perceptron_weights.npy', best_weights)
```

```python
# Load the saved weights with minimum error
w = np.load('perceptron_weights.npy)

# Create a new perceptron instance with the best weights
perceptron_test = Perceptron(numb_fe=6)
perceptron_test.weights = w

# Calculate accuracy on the testing set
corr_prediction = 0
total = len(X_test)

for i in range(total):
    prediction = perceptron_test.predict(X_test[i])
    if prediction == y_test[i]:
        corr_prediction += 1

accuracy = corr_prediction / total
print("Accuracy:", accuracy)

```
