```python
#checking outlier in Age
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers

outliers = find_outliers_IQR(df['Age'])

print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))
```

```python
#find mean, median, mode value for Age
df_temp = df
median_val = df_temp['Age'].median()
mean_val = df_temp['Age'].mean()
mode_val = df_temp['Age'].mode()

print("Median Value of age: ",median_val)
print("Mean Value of age: ",mean_val)
print("Mode Value of age: ",mode_val,"\n\n")

#replacing null value with median value
df_temp['Age'] = df_temp['Age'].fillna(median_val)
df_temp.isnull().sum()
```

```python
# Drop Cabin column
df_temp = df

df_temp.pop('Cabin')
df_temp.head()
df_temp.isnull().sum()
```
