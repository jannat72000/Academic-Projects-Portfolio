# Annotate
Annotate the point xy with text text.

In the simplest form, the text is placed at xy.

Optionally, the text can be displayed in another position xytext. An arrow pointing from the text to the annotated point xy can then be added by defining arrowprops.

\

Learn More:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot
plt.plot([1, 2, 3, 4])

# Label
plt.ylabel('Some Numbers')

# Set x limits
plt.xlim(0,3.0)

# Set y limits
plt.ylim(1.0,4.0)

# Set x ticks
plt.xticks([0,1,2,3],
           ['$0$', '$1$', '$2$','$3$'])

# plt.xticks([0,1,2,3],
#            ['$-\pi$', r'$-\frac{\pi}{2}$', '$0$', '$+\pi/2$'])

# Using mathtext to get Unicode Characters by wrapping the format specifier with $

# Set y ticks
plt.yticks([1,2,3,4],
           ['$1$', '$2$', '$3$','$4$'])


# Dotted line plot
plt.plot([2, 2], [1, 3], color = 'red', linestyle = "--", linewidth = 1.0)

# Annotate
plt.annotate(r'$y = 3$', xy = (2, 3), xytext = (+50, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))

# Title
plt.title("Line")

plt.show()
```

```python
# Code Here

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ax = np.linspace(-2*np.pi, 2*np.pi)
plt.ylim(-1,1)

plt.xlim(-2*np.pi, 2*np.pi)

plt.yticks([-1,0,1],
           ['$-1$', '$0$', '$+1$'])

plt.xticks([-2*np.pi, -3*np.pi /2 , - np.pi , -np.pi /2 , 0,   np.pi /2,np.pi,3*np.pi /2, 2*np.pi],
           ['$-2\pi$', r'$-\frac{3\pi}{2}$','$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$','$\pi$', r'$\frac{3\pi}{2}$','$2\pi$'])


cos = np.cos(ax)

plt.plot(ax,cos, color = 'red')
plt.title("Cosine Curve")

plt.plot([np.pi/3, np.pi/3], [-1, 0.5 ], color = 'red', linestyle = "--", linewidth = 1.0)

plt.annotate(r'$cos(\frac{\pi}{3}) = 1/2$', xy = (np.pi/3 , 0.5), xytext = (+30, -25), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
```

# Exercise 2:

Draw a box plot from "adult.data" with age column and grouping by sex column.

Use df.boxplot() for plotting.

Use Parameters:

Grid should be false.

Also, give appropriate figure size.

Now use Annotate to pinpoint:


*   Median (With Value)
*   Upper Quartile (Q3) (With Value)
*   Lower Quartile (Q1) (With Value)
*   Upper Whisker (With Value)
*   Lower Whisker (With Value)

Also, pinpoint one outlier (If any).




```python
# Code Here

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/adult.data', header=None)
df.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain',
              'capital-loss','hours-per-week','native-country', 'salary']

a = df.boxplot(by= 'sex', column = 'age', grid = False, figsize = (14,12))

gender = df.groupby(['sex'])['age']
m = gender.median()

#female median
a.annotate('$median = %.2f $'%m[0], xy = (1 , m[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#male median
a.annotate('$median = %.2f$'%m[1] , xy = (2 , m[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#quatile
lower_q = df.groupby(['sex'])['age'].quantile(0.25)

#female q1
a.annotate('$Lower Quatile = %.2f $' % lower_q[0], xy = (1 , lower_q[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
#male q1
a.annotate('$Lower Quatile = %.2f $' %lower_q[1] , xy = (2 , lower_q[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#q3
upper_q = df.groupby(['sex'])['age'].quantile(0.75)

#female q3
a.annotate('$Upper Quatile = %.2f $' %upper_q[0], xy = (1 , upper_q[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
#male q3
a.annotate('$Upper Quatile$ = %.2f ' %upper_q[1], xy = (2 , upper_q[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

max = df.groupby(['sex'])['age'].max()

min = df.groupby(['sex'])['age'].min()

IQR = upper_q - lower_q

Lower_Extreme = lower_q - (1.5 * IQR)

Upper_Extreme = upper_q + (1.5 * IQR)

#female_upper_whisker
if max[0] < Upper_Extreme[0]:
  a.annotate('$Upper Whisker$ = %.2f ' %max[0], xy = (1 , max[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
if max[0] > Upper_Extreme[0]:
  a.annotate('$Upper Whisker$ = %.2f ' %Upper_Extreme[0], xy = (1 , Upper_Extreme[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#male_upper_whisker
if max[1] < Upper_Extreme[1]:
  a.annotate('$Upper Whisker$ = %.2f ' %max[1], xy = (2 , max[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
if max[1] > Upper_Extreme[1]:
  a.annotate('$Upper Whisker$ = %.2f ' %Upper_Extreme[1], xy = (2 , Upper_Extreme[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#female_lower_whisker
if min[0] > Lower_Extreme[0]:
  a.annotate('$Lower Whisker$ = %.2f ' %min[0], xy = (1 , min[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
if min[0] < Lower_Extreme[0]:
  a.annotate('$Lower Whisker$ = %.2f ' %Lower_Extreme[0], xy = (1 , Lower_Extreme[0]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

#male_lower_whisker
if min[1] > Lower_Extreme[1]:
  a.annotate('$Lower Whisker$ = %.2f ' %min[1], xy = (2 , min[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))
if min[1] < Lower_Extreme[1]:
  a.annotate('$Lower Whisker$ = %.2f ' %Lower_Extreme[1], xy = (2 , Lower_Extreme[1]), xytext = (+25, -15), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .4"))

```

```python
# Code Here

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/adult.data')
df.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain',
              'capital-loss','hours-per-week','native-country', 'salary']

fig, axes = plt.subplots(nrows = 2, ncols = 2 , figsize = (12,9))

ax = df.plot('age', 'education-num',
                      c='Blue', linestyle = ':', linewidth = 3, ax = axes[0,0])

ax.set_title('Age and education-num')

ax = df.plot('age', 'capital-gain',
                      c='Blue', linestyle = ':', linewidth = 3, ax = axes[0,1])

ax.set_title('Age and Capital gain')

ax = df.plot('age', 'capital-loss',
                      c='green', linestyle = ':', linewidth = 3, ax = axes[1,0])

ax.set_title('Age and capital-loss')

ax = df.plot('age', 'hours-per-week',
                      c='black', linestyle = ':', linewidth = 3, ax = axes[1,1])

ax.set_title('Age and hours-per-week')


```
