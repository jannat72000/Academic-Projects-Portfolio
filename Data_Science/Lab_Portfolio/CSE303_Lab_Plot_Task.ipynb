# Exercise 1:
Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –

Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

Line width should be 3

image.png



```python
# Code Here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/CSE303/company_sales_data.csv')

ax = df.plot("month_number", "total_units", linestyle = 'dashed', linewidth = 3,color = 'red', marker = 'o', mfc='black', label ="Profit data of last year" )
ax.set_xlabel('Month Number')
ax.set_ylabel('profit in dollar')
ax.set_title('Company sales data of last year')

plt.ylim(10000,50000)

label = df.month_number
plt.xticks(label)
plt.show()

```

# Exercise 2:
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.

image.png

```python
# Code Here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/company_sales_data.csv')

ax = df.plot('month_number',['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'],
             linewidth = 3, marker = 'o', label =["Face Cream Sales Data","facewash Sales Data",
            "Tooth Paste Sales Data" ,"bathingsoap Sales Data","shampoo Paste Sales Data"
            ,"moisturizer Paste Sales Data"])

ax.set_xlabel('Month Number')
ax.set_ylabel('sales unit in number')
ax.set_title('sales data')

plt.ylim(1000,18000)

label = df.month_number
plt.xticks(label)

plt.show()

```

# Exercise 3:
Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

```python
# Code Here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/company_sales_data.csv')

df = pd.DataFrame({'face wash sales data': df.facewash ,'face creame sales data': df.facecream}, index=df.month_number  )

ax = df.plot.bar(rot=0, color={"face wash sales data": "blue", "face creame sales data": "red"})

ax.set_xlabel('Month Number')
ax.set_ylabel('sales unit in number')
ax.set_title('face wash and face creame sales data')

plt.grid(linestyle='dashed', linewidth=1)


```

# Exercise 4:
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

image.png

```python
# Code Here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/company_sales_data.csv')

ax = plt.pie([sum(df.facecream), sum(df.facewash), sum(df.toothpaste), sum(df.bathingsoap),sum(df.shampoo), sum(df.moisturizer)],
             autopct = '%1.1f%%', labels = ["FaceCream ","facewash ","ToothPaste " ,"bathingsoap ","shampooPaste" ,"moisturizer Paste"])

plt.legend(loc="center left", bbox_to_anchor=(0.8, 0, 0.5, 0.5))
plt.title('sales data')

plt.show()
```

# Exercise 5:
Plot Sine and Cosine Graph.

The plot graph should look like this.

image.png

```python
# Code Here

import matplotlib.pyplot as plt
import numpy as np

ax = np.linspace(-np.pi, np.pi)

cos = np.cos(ax)
sin = np.sin(ax)

plt.figure(figsize=(7,5))

plt.plot(ax,cos, color = 'blue')
plt.plot(ax,sin, color = 'green')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sine and Cosine Graph')
plt.legend(['cosine','sine'])

plt.show()
```
